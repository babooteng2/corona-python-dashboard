```python
import plotly.express as px
from data import countries_df

print(countries_df.head())

fig = px.scatter_geo(
    countries_df,
    hover_name="Country_Region",
    color="Confirmed",    
    locations="Country_Region",
    locationmode="country names",
    hover_data={
        "size": "Confirmed",
        "Confirmed": ":,.2f",
        "Deaths": ":,.2f",
        "Recovered": ":,.2f",
        "Country_Region": False
    })
fig.show()
```
